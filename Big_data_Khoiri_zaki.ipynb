```python
import pandas as pd

# Daftar path file dataset
file_paths = [
    "/content/drive/My Drive/Colab Notebooks/tutorial/bigdata/dataset/Luas Panen, Produktivitas, dan Produksi Padi Menurut Provinsi, 2018.csv",
    "/content/drive/My Drive/Colab Notebooks/tutorial/bigdata/dataset/Luas Panen, Produktivitas, dan Produksi Padi Menurut Provinsi, 2019.csv",
    "/content/drive/My Drive/Colab Notebooks/tutorial/bigdata/dataset/Luas Panen, Produktivitas, dan Produksi Padi Menurut Provinsi, 2020.csv",
    "/content/drive/My Drive/Colab Notebooks/tutorial/bigdata/dataset/Luas Panen, Produktivitas, dan Produksi Padi Menurut Provinsi, 2021.csv",
    "/content/drive/My Drive/Colab Notebooks/tutorial/bigdata/dataset/Luas Panen, Produktivitas, dan Produksi Padi Menurut Provinsi, 2022.csv"
]


# Dictionary untuk mengganti nama kolom
ganti_nama = {
    "Luas Panen Tanaman Padi (ha) (ha)": "luas panen",
    "Produktivitas Tanaman Padi (ku/ha) (ku/ha)": "produktivitas",
    "Rekap Produksi Padi (ton) (ton)": "rekap produksi"
}

# Membaca dan menampilkan dataset dengan kolom yang sudah diubah
for i, file_path in enumerate(file_paths, start=2018):
    dataset = pd.read_csv(file_path)

    # Tampilkan nama kolom asli
    print(f"\nNama Kolom Asli untuk Dataset Tahun  {i}:")
    print(dataset.columns)

    # Ganti nama kolom sesuai mapping
    dataset = dataset.rename(columns=ganti_nama)

    print(f"\nDataset Tahun {i} (10 Baris Pertama dengan Kolom yang Diubah):")
    display(dataset.head(10))  # Menampilkan 10 baris pertama dengan kolom yang sudah diubah
```

```python
# List untuk menyimpan dataset setelah diubah
semua_data = []

# Membaca setiap dataset, mengganti nama kolom, dan menambahkannya ke list
for file_path in file_paths:
    dataset = pd.read_csv(file_path)
    dataset = dataset.rename(columns=ganti_nama)  # Ganti nama kolom
    semua_data.append(dataset)  # Tambahkan dataset ke dalam list

# Menggabungkan semua dataset
gabungan_data = pd.concat(semua_data, ignore_index=True)

# Menampilkan dataset gabungan
print("Dataset Gabungan:")
display(gabungan_data)  # Menampilkan dataset gabungan
```

```python
# Menampilkan informasi dari dataset gabungan
gabungan_data.info()
```

```python
# Menampilkan statistik deskriptif dataset gabungan
print(gabungan_data.describe())
```

```python
# Menampilkan nilai null atau missing values dalam dataset gabungan
print(gabungan_data.isnull().sum())
```

```python
# Menampilkan baris berapa saja memiliki nilai null
cek_null = gabungan_data[gabungan_data.isnull().any(axis=1)]
print(cek_null)
```

```python
# Melakukakan drop untuk data yang ada pada kolom provinsi
buang_provinsi_null = cek_null['Provinsi'].unique()
data_filter = gabungan_data[~gabungan_data['Provinsi'].isin(buang_provinsi_null)]
```

```python
# Menampilkan data setelah prepossessing
print(data_filter.isnull().sum())
```

```python
# Menampilkan isi data yang telah di prepossessing dengan rapi
display(data_filter)
```

```python
# Membuang setiap kata dalam kolom provinsi dengan key word indonesia
data_filter = data_filter[data_filter['Provinsi'] != 'Indonesia']
```

```python
# Menampilkan isi data yang telah di prepossessing dengan rapi
display(data_filter)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Gabungkan data provinsi dan tahun lalu menjumlahkan produksinya
gabungan_data_2 = data_filter.groupby(['Provinsi', (data_filter.index // 34) + 2017])['rekap produksi'].sum().reset_index()
gabungan_data_2.rename(columns={gabungan_data_2.columns[1]:'Tahun'},inplace=True)
gabungan_data_2['Tahun'] = gabungan_data_2['Tahun'] +1  # Add 1 to the 'Tahun' column

# Filter data tahun 2018-2022
filter_gabungan_data = gabungan_data_2[(gabungan_data_2['Tahun'] >= 2018) & (gabungan_data_2['Tahun'] <= 2022)]

# Membuat visualisasi Line Chart
plt.figure(figsize=(12, 6))
sns.lineplot(data=filter_gabungan_data, x='Tahun', y='rekap produksi', hue='Provinsi')
plt.title('Tren Produksi Padi per Provinsi (2018-2022)')
plt.xlabel('Tahun')
plt.ylabel('Total Produksi Padi (ton)')
plt.xticks(filter_gabungan_data['Tahun'].unique())
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Mneghitung rata-rata produksi tiap provinsi
rata2_produksi = data_filter.groupby('Provinsi')['produktivitas'].mean().reset_index()

# Melakukan pengurutan data berdasarkan rata-rata produksi dari tiap provinsi dengan descending
rata2_produksi = rata2_produksi.sort_values(by=['produktivitas'], ascending=False)


# Visualisasi bar plot
plt.figure(figsize=(12, 6))
sns.barplot(data=rata2_produksi, x='Provinsi', y='produktivitas')
plt.title('Rata-rata Produktivitas Padi per Provinsi')
plt.xlabel('Provinsi')
plt.ylabel('Produktivitas (ku/ha)')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```
